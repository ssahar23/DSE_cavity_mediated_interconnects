Proposed figure of merit
$$
\Gamma =  \frac{\alpha \times \text{Efficiency}}{(1-\alpha) \times (\text{Latency} \times \text{Infidelity})}
$$

such that:
$$
\text{Efficiency} \propto P_s = (\frac{\kappa_{ex}}{\kappa}) (\frac{2C}{1+2C}) (1 - \frac{I_g'}{\kappa C})
$$

$$
\text{Infidelity} \propto \sqrt\frac{\kappa \gamma}{g^2}
$$

$$
\text{Latency} \propto \frac{Q}{\omega_c} = \frac{\Delta^2 + \gamma^2}{2 g^2 \gamma + \kappa(\Delta^2 + \gamma^2)}
$$

```python
from matplotlib.colors import to_hex, to_rgb
import colorsys


def adjust_color_moderate(hex_color, lightness_increase=0.3, saturation_increase=0.1):
    # Convert hex to RGB
    rgb = to_rgb(hex_color)
    # Convert RGB to HSL
    h, s, l = colorsys.rgb_to_hls(*rgb)
    # Adjust lightness and saturation, ensuring they do not exceed 1
    l = min(l + lightness_increase, 1.0)
    s = min(s + saturation_increase, 1.0)
    # Convert back to RGB
    new_rgb = colorsys.hls_to_rgb(h, s, l)
    # Convert RGB to hex
    return to_hex(new_rgb)


# Brighten the palette with moderate adjustment
moderate_bright_palette = [
    adjust_color_moderate(color) for color in color_palette_bright]
moderate_bright_palette


sns.color_palette(moderate_bright_palette)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Constants
alpha = 0.5
kappa_ex = 1e6 # External cavity decay rate
Ig_prime = 1 # System imperfections
Delta = 2e9  # Frequency detuning, 2 GHz

# Varying g from 10^3 to 10^9
g_values = np.logspace(3, 9, 100)

# Prepare kappa and gamma values for the plots
kappa_gamma_pairs = [(10**4, 10**4), (10**5, 10**5), (10**6, 10**6)]

plt.figure(figsize=(10, 7))

# Find the inflection point by calculating the second derivative
def find_inflection_point(g_values, FM):
    # Compute the second derivative
    second_derivative = np.gradient(np.gradient(FM, g_values), g_values)
    # Find where the second derivative crosses zero
    inflection_index = np.where(np.diff(np.sign(second_derivative)))[0]
    # Return the inflection point, if it exists
    return g_values[inflection_index[0]] if inflection_index.size > 0 else None

for kappa, gamma in kappa_gamma_pairs:
    # Calculate cooperativity C for each kappa and gamma pair
    C = (g_values**2) / (2*(kappa * gamma))
    
    # Calculate Efficiency (P_s)
    P_s = (kappa_ex / kappa) * (2 * C / (1 + 2 * C)) * (1 - Ig_prime / (kappa * C))
    
    # Calculate Infidelity
    infidelity = np.sqrt(kappa * gamma / g_values**2)
    
    # Calculate Latency using corrected variables
    latency = ((Delta**2 + gamma**2) / (2 * (g_values**2) * gamma + kappa * (Delta**2 + gamma**2)))

    # Calculate FM
    FM = (alpha * P_s) / ((1 - alpha) * (latency * infidelity))


    # Find and plot the inflection point
    if kappa == 10**4:
        plt.plot(g_values, np.log10(FM),color='blue', label=f'$\kappa = \gamma = 10^4$ Hz')
        inflection_point = find_inflection_point(g_values, FM)
        if inflection_point:
            plt.axvline(x=inflection_point, color='blue', linestyle='--', lw=1, alpha=0.7)
    if kappa == 10**5:
        plt.plot(g_values, np.log10(FM),color='orange', label=f'$\kappa = \gamma = 10^5$ Hz')
        inflection_point = find_inflection_point(g_values, FM)
        if inflection_point:
            plt.axvline(x=inflection_point, color='orange', linestyle='--', lw=1, alpha=0.7)
    if kappa == 10**6:
        plt.plot(g_values, np.log10(FM),color='green', label=f'$\kappa = \gamma = 10^6$ Hz')
        inflection_point = find_inflection_point(g_values, FM)
        if inflection_point:
            plt.axvline(x=inflection_point, color='green', linestyle='--', lw=1, alpha=0.7)

plt.xscale('log')

plt.xlabel('Coupling Strength g [Hz]', fontsize=20)
plt.ylabel('(Log) Figure of Merit $\Gamma$', fontsize=20)
plt.tick_params(axis='both', labelsize=15)
plt.legend(fontsize=15)
plt.grid(True)
plt.savefig(f'./figures/sensitivity_analysis_1.pdf', format="pdf", dpi=400)
```
